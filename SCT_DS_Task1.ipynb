```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel("food_order.xls")
df.head()
```

```python
day_counts = df['day_of_the_week'].value_counts()

plt.figure()
plt.bar(day_counts.index, day_counts.values , edgecolor='black')
plt.xlabel("Day of the Week")
plt.ylabel("Number of Orders")
plt.title("Distribution of Orders by Day of the Week")
plt.show()
```

```python
plt.figure()
plt.hist(df['delivery_time'], bins=10, edgecolor='black')
plt.xlabel("Delivery Time (minutes)")
plt.ylabel("Frequency")
plt.title("Distribution of Delivery Time")
plt.show()
```
